```python
# I will try to retrieve one instance of the file in a list of dictionaries
import json
import pandas as pd
records = [json.loads(line) for line in open('01_data/usa.gov_click_data_1.json')]
# Print the first occurance
records[3] 
```

```python
df_records = pd.DataFrame(records)
```

```python
df_rec = pd.DataFrame()
```

```python
df_rec['web_browser'] = df_records['a'].str.split().str[0].str.strip() 
df_rec['operating_sys'] = df_records['a'].str.extract(r"\((.*?)\)", expand=False) 
df_rec['from_url'] = df_records['r'] 
df_rec['to_url'] = df_records['u'] 
df_rec['city'] = df_records['cy'] 
df_rec['longitude'] = df_records['ll'].str.get(0) 
df_rec['latitude'] = df_records['ll'].str.get(1) 
df_rec['time_zone'] = df_records['tz'] 
df_rec['time_in'] = df_records['t'] 
df_rec['time_out'] = df_records['hc'] 
df_rec['time_in'] = pd.to_datetime(df_rec['time_in'],unit='ms') 
df_rec['time_out'] = pd.to_datetime(df_rec['time_out'],unit='ms') 
#df_rec.to_csv(r'target/usa.gov_click_data_2.csv',header=True, index = False) 
```
